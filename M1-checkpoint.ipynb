```python
df1 = df.copy()
```

```python
import inflection
import re
```

```python
cols_old = df.columns
```

```python
# Function to remove special characters and spaces
def clean_column_name(name):
    name = re.sub(r'[^\w\s]', '', name)  # Remove special characters
    name = name.replace(' ', '_')  # Replace spaces with underscores
    return name
snakecase = lambda x: inflection.underscore(x)
# Apply the cleaning function and then the snake case conversion
cols_cleaned = list(map(lambda x: snakecase(clean_column_name(x)), cols_old))
```

```python
cols_cleaned
```

```python
# Replace the column names in the DataFrame
df1.columns = cols_cleaned
```
